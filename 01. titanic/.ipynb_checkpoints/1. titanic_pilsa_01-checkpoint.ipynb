### 1-1. (1) Percent로 결측치 확인

```python
# '>' 10칸 할당 + 오른쪽 정렬
for col in df_train.columns:
    msg = 'columns: {:>10}\t Percent of NaN value: {:.2f}%'.format(col, 100*(df_train[col].isnull().sum() / df_train[col].shape[0]))
    print(msg)
```

```python
for col in df_test.columns:
    msg = 'columns: {:>10}\t Percent of NaN value: {:.2f}%'.format(col, 100*(df_test[col].isnull().sum() / df_test[col].shape[0]))
    print(msg)
```

```python
# 위에 것 함수화 하기
def msg(df):
    for col in df.columns:
        print ('columns: {:>10}\t Percent of NaN value: {:.2f}%'.format(col, 100*(df[col].isnull().sum() / df[col].shape[0])))
```

```python
# 함수 적용
msg(df_test)
```

### 2.4 Age

```python
print('제일 나이 많은 탑승객 : {:.1f} Years'.format(df_train["Age"].max()))
print('제일 어린 탑승객 : {:.1f} Years'.format(df_train["Age"].min()))
print('탑승객 평균 나이 : {:.1f} Years'.format(df_train["Age"].mean()))
```

```python
# 생존에 따른 Age의 histogram


fig, ax = plt.subplots(1, 1, figsize = (9, 5))

sns.kdeplot(df_train[df_train["Survived"] == 1]["Age"], ax = ax)
sns.kdeplot(df_train[df_train["Survived"] == 0]["Age"], ax = ax)

plt.legend(["Survived == 1", "Survived == 0"])
plt.show()
```

- 생존자 중 나이가 어린 경우가 더 많음

```python
plt.figure(figsize = (8,6))

# df_train["Age"][df_train["Pclass"] == 1] 이거 중에 결측치도 존재하는데

df_train["Age"][df_train["Pclass"] == 1].plot(kind = 'kde')
df_train["Age"][df_train["Pclass"] == 2].plot(kind = 'kde')
df_train["Age"][df_train["Pclass"] == 3].plot(kind = 'kde')

plt.xlabel("Age")
plt.title('Age Distribution within classes')
plt.legend(["1st Class", "2nd Class", "3rd Class"])
```

```python
cummulate_survival_ratio = []

# df_train[df_train["Age"] < 4]["Survived"] : age가 4보다 작은 데이터의 Survived

for i in range(1,80):
    cummulate_survival_ratio.append(df_train[df_train["Age"] < i]["Survived"].sum() / 
                                   len(df_train[df_train["Age"] < i]["Survived"]))
    # 1의 합 / 전체

plt.figure(figsize = (7,7))

plt.plot(cummulate_survival_ratio)
plt.title('Survival rate change depending on range of Age', y = 1.02)
plt.ylabel('Survival rate')
plt.xlabel('Range of Age(0~x)')
plt.show()
    
```

- 나이가 어릴 수록 생존률이 확실히 높음
- 나이가 중요한 feature로 쓰일 수 있음

### 2.8 Fare

```python
f, ax = plt.subplots(1, 1, figsize = (8,8 ))
g = sns.distplot(df_train['Fare'], color = 'b', label = 'Skewness : {:.2f}'.format(df_train["Fare"].skew()), ax=ax)

g= g.legend(loc='best')
```

- Fare은 탑승요금
- distribution이 매우 비대칭 -> 이대로 모델에 넣어준다면 모델이 잘못 학습할 수 도 있음.
- 몇 개 없는 outlier에 대해서 너무 민감하게 반응할 시, 실제 예측 시에 좋지 못한 결과를 부를 수 있음
<br>

- outlier의 영향을 줄이기 위해 Fare에 log 취하기

```python
df_test.loc[df_test.Fare.isnull(), "Fare"] = df_test["Fare"].mean() # Fare가 NaN 인 데이터를 평균값으로 치환

df_train["Fare"] = df_train["Fare"].map(lambda i : np.log(i) if i > 0 else 0)
df_test["Fare"] = df_test["Fare"].map(lambda i : np.log(i) if i > 0 else 0)
```

```python
f, ax = plt.subplots(1, 1, figsize = (8,8))

g = sns.distplot(df_train["Fare"], color = "b", label = "Skewness : {:.2f}".format(df_train["Fare"].skew()), ax =ax)

g = g.legend(loc = 'best')
```

- log 를 취하니, 비대칭성이 많이 사라졌다.
- 느낀점 : 변수의 분포를 확인하고 log를 통해 비대칭성을 변화시키기
